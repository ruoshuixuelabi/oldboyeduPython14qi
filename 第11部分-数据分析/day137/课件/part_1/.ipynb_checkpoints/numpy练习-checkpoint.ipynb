```python
plt.imshow(img[::-1])  #将上面图片上下翻转
plt.imshow(img[:,::-1])  #左右翻转
plt.imshow(img[::-1,::-1])  #上下左右翻转
plt.imshow(img[:,:,::-1])  #rgb颜色翻转
```

```python
pic = np.random.randint(0,255,(200,1,3))

pic = np.concatenate(tuple([pic]*200),axis=1)
plt.imshow(pic)
```

```python
pic = np.random.randint(0,255,(200,1,1))

pic = np.concatenate(tuple([pic]*200),axis=1)
pic = np.concatenate(tuple([pic]*3),axis=2)

plt.imshow(pic)
```

```python
plt.imshow(img)
```

```python
new_img = np.vstack((img,img))
plt.imshow(new_img)
```

```python
new_img = np.hstack((img,img))
plt.imshow(new_img)
```

```python
new_img = np.split(img,[0,400],axis=0)  #axis=0横切
plt.imshow(new_img[1])
```

```python
new_img = np.split(img,[0,400],axis=1)  #axis=1竖切
plt.imshow(new_img[1])
```
