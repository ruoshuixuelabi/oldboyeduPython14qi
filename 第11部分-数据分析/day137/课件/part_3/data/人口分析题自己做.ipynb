```python
import numpy as np
import pandas as pd
```

```python
abb = pd.read_csv('./state-abbrevs.csv')
```

```python
area = pd.read_csv('./state-areas.csv')
```

```python
popu = pd.read_csv('./state-population.csv')
```

```python
data=pd.merge(abb,popu,left_on='abbreviation',right_on='state/region',how='outer')
data
```

```python
new_data = data.drop(columns='abbreviation')
new_data
```

```python
new_data.isnull().any(axis=0)  #查存在缺失数据的列
```

```python
#找到有哪些state/region的state的值为NaN，进行去重操作
#unique()去重
index = new_data['state'][new_data['state'].isnull()].index  #空值的index列表
s = new_data.iloc[index]['state/region']
s.unique() #去重操作
```

```python
#为找到的这些state/region的state项补上正确的值，从而去除掉state这一列的所有NaN
index_list = new_data['state'][new_data['state'].isnull()].index
new_data.iloc[index_list]['state/region'].unique()
```

```python
con = new_data['state/region']=='USA'
new_data.loc[con,'state']='UnitedState'
new_data
```
