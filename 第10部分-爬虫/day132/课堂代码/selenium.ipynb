```python
from selenium import webdriver
from time import sleep
```

```python
bro = webdriver.Chrome(executable_path=r'C:\Users\Administrator\Desktop\爬虫+数据\day04\chromedriver.exe')

#请求的发送
bro.get(url='https://www.baidu.com')
sleep(3)
#根据find系列的函数定位到指定的标签
my_input = bro.find_element_by_id('kw')

#向标签中录入指定的数据
my_input.send_keys('人民币')
sleep(3)

my_button = bro.find_element_by_id('su')
my_button.click()
sleep(3)
#获取当前浏览器显示的页面的页面源码
page_text = bro.page_source
print(page_text)
bro.quit()
```

```python
bro = webdriver.Chrome(executable_path=r'C:\Users\Administrator\Desktop\爬虫+数据\day04\chromedriver.exe')
url = 'https://movie.douban.com/typerank?type_name=%E7%88%B1%E6%83%85&type=13&interval_id=100:90&action='
#请求的发送
bro.get(url=url)

js = 'window.scrollTo(0,document.body.scrollHeight)'
bro.execute_script(js)
sleep(2)
bro.execute_script(js)
sleep(2)
bro.execute_script(js)
sleep(2)

page_text = bro.page_source

with open('./douban.html','w',encoding='utf-8') as fp:
    fp.write(page_text)
    
bro.quit()
```

```python
#登录qq空间
bro = webdriver.Chrome(executable_path=r'C:\Users\Administrator\Desktop\爬虫+数据\day04\chromedriver.exe')
url = 'https://qzone.qq.com/'
#请求的发送
bro.get(url=url)
sleep(1)
#定位到指定的iframe
bro.switch_to.frame('login_frame')

bro.find_element_by_id('switcher_plogin').click()
sleep(1)

userName = bro.find_element_by_id('u')
userName.send_keys('916715469')

passWord = bro.find_element_by_id('p')
passWord.send_keys('zm916715469')
sleep(1)
bro.find_element_by_id('login_button').click()
sleep(1)
js = 'window.scrollTo(0,document.body.scrollHeight)'
bro.execute_script(js)
sleep(2)
bro.execute_script(js)
sleep(2)
bro.execute_script(js)
sleep(2)
page_text = bro.page_source
sleep(5)

#解析
tree = etree.HTML(page_text)
div_list = tree.xpath('//div[@class="f-info qz_info_cut"] | //div[@class="f-info"]')
for div in div_list:
    text = div.xpath('.//text()')
    text = "".join(text)
    print(text)
    
bro.quit()
```
